```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

We will generate our training data randomly
The real function is 4 + 3x1 + 4x2 + (some random noise)
We can just use a dot product for easier computatation


```
import numpy as np
np.random.seed(42)

xs = 2 * np.random.rand(100, 2)

ys = 4 + np.dot(xs, np.array([[3], [4]])) + np.random.rand(100, 1)
```

```
x1 = np.array([d[0] for d in xs])
x2 = np.array([d[1] for d in xs])
ax = plt.axes(projection='3d')
ax.scatter(x1, x2, ys, alpha=1, color='r', s=10)
```

Looks kind of like theres clearly a plane that describes the function


```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# split the data into training and test sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]
```

We will start with two random numbers for our initial weights and [0,0] for the bias

```
# initial value for weight w and bias b
w = np.random.randn(2)
b = np.zeros(2)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = np.dot(w, train_xs[i]) + b
    
    grad_w = (y_pred - train_ys[i]) * train_xs[i]
    grad_b = (y_pred - train_ys[i])
    
    w -= lr * grad_w
    b -= lr * grad_b
```
