```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

Local Maximum


```
X = np.linspace(-1.2, 1.2)
Y = np.linspace(-1.2, 1.2)

def hills(X, Y):
  return np.cos(X) * np.cos(y) * (np.exp(-X**2 * -y**2))

x, y = np.meshgrid(X, Y)
z = hills(x, y)

ax = plt.axes(projection='3d')
ax.scatter([x[25,25]], [y[25,25]], [z[25,25]], alpha=1, color='r', s=100)
ax.plot_surface(x, y,z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_title('surface');
```

As you can see (where the little red ball is), there is a maximum at the top of the hill, but all of the corners go up to inifinity, so the center is a local maximum.

Local Minimum


```
X = np.linspace(-1.2, 1.2)
Y = np.linspace(-1.2, 1.2)

def reverse_hills(X, Y):
  return (np.cos(X) * np.cos(y) * (np.exp(-X**2 * -y**2))) * -1

x, y = np.meshgrid(X, Y)
z = reverse_hills(x, y)

ax = plt.axes(projection='3d')
ax.scatter([x[25,25]], [y[25,25]], [z[25,25]], alpha=1, color='r', s=100)
ax.plot_surface(x, y,z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_title('surface')
```

If we take the negation of the previous function, we now get a creepy looking thing that looks like some alien spider. Where I presume its mouth for consuming human flesh would  be, we find a local minimum. Similar to before, it's legs stretch down to infinity (as it needs to see it's victims from high in the sky), making this point in the middle a local minimum

```
X = np.linspace(-1, 1)
Y = np.linspace(-1, 1)

def stairs(X, Y):
  return (np.sign(-.65-X)+np.sign(-.35-X)+np.sign(-.05-X)+np.sign(.25-X)+np.sign(.55-X))/7

x, y = np.meshgrid(X, Y)
z = stairs(x, y)

ax = plt.axes(projection='3d')
ax.scatter([x[25,25]], [y[25,25]], [z[25,25]], alpha=1, color='r', s=100)
ax.plot_surface(x, y,z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.set_title('surface');
```

As we can see, the gradient at the point is zero (flat), but it is netiehr a minimum nor a maximum
